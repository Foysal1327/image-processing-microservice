```python
import requests
import json
import matplotlib.pyplot as plt
import numpy as np
import tifffile as tiff
from PIL import Image
from io import BytesIO
```

```python
import os

file_path = "data/test_image.tif"
print("File exists:", os.path.exists(file_path))
```

```python
import numpy as np
import tifffile as tiff

dummy_image = np.random.randint(0, 256, (100, 100, 10, 5, 3), dtype=np.uint8)


os.makedirs("data", exist_ok=True)

tiff.imwrite("data/test_image.tif", dummy_image)
print("Test image created successfully.")
```

```python
file_path = "data/test_image.tif"

with open(file_path, "rb") as file:
    response = requests.post("http://127.0.0.1:5000/upload", files={"file": file})

print(response.json())
```

```python
params = {"file_path": "data/test_image.tif"}
response = requests.get("http://127.0.0.1:5000/metadata", params=params)

metadata = response.json()
print(json.dumps(metadata, indent=4))
```

```python
params = {
    "file_path": "data/test_image.tif",
    "z": 5,
    "time": 2,
    "channel": 1  
}
response = requests.get("http://127.0.0.1:5000/slice", params=params)


slice_img = Image.open(BytesIO(response.content))

plt.figure(figsize=(6, 6))
plt.imshow(slice_img, cmap="gray")
plt.title("Extracted Image Slice")
plt.axis("off")
plt.show()
```

```python
data = {
    "file_path": "data/test_image.tif",
    "components": 3
}
response = requests.post("http://127.0.0.1:5000/analyze", json=data)

pca_result = response.json()
print(json.dumps(pca_result, indent=4))
```

```python
params = {"file_path": "data/test_image.tif"}
response = requests.get("http://127.0.0.1:5000/statistics", params=params)

stats = response.json()
print(json.dumps(stats, indent=4))
```

```python
# Define API request parameters
params = {
    "file_path": "data/test_image.tif",  # The uploaded image file
    "channel": 0,  # Select a specific spectral channel
    "k": 3  # Number of clusters for K-Means
}

response = requests.get("http://127.0.0.1:5000/segment/kmeans", params=params)

if response.status_code == 200:
    print("K-Means segmentation successful!")
    segmented_file = response.json()["file_path"]
    print("Segmented image saved at:", segmented_file)
else:
    print("Error:", response.json())
```

```python
# Load the segmented image
segmented_image = tiff.imread(segmented_file)

# Display the segmented image
plt.figure(figsize=(6, 6))
plt.imshow(segmented_image, cmap="gray")
plt.title("K-Means Segmentation Output")
plt.axis("off")
plt.show()
```
